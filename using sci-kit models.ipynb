```python
x=d.drop(columns=['BI_RADS','Severity'])
```

```python
x.head()
```

```python
y=d.drop(columns=['BI_RADS','Age','Shape','Margine','Density'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
x_test.head()
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
classification_report(y_test,y_pred)
```

```python
from sklearn import svm
clf1 = svm.SVC(gamma='scale')
clf1.fit(x_train,y_train)  
```

```python
y_pred2=clf1.predict(x_test)
```

```python
clf1.support_vectors_
```

```python
print(classification_report(y_test,y_pred2))
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf2 = KNeighborsClassifier(n_neighbors=10)
clf2.fit(x_train,y_train)
y_pred3=clf2.predict(x_test)
```

```python
print(classification_report(y_test,y_pred3))
```

```python
import seaborn as sns
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier=Sequential()
classifier.add(Dense(4,activation='relu',input_dim=4))
classifier.add(Dense(1,activation='sigmoid'))
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=["accuracy"])
```
